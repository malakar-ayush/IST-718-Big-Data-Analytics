```python
# ==============================================================================
# Listing out the composite files generated
# ==============================================================================

ValuesComposit_fileList = []
ModelsComposit_fileList = []

for file in os.listdir(tempDir):
    if file.startswith("ValuesComposit_"):
        ValuesComposit_fileList.append(file)
    if file.startswith("ModelsComposit_"):
        ModelsComposit_fileList.append(file)
```

```python
# ==============================================================================
# Plotting Values Composite DataFrames
# ==============================================================================

count = 0
for file in ValuesComposit_fileList:
    count += 1
    print(count, ': '+tempDir+'/'+file)
    df = pd.read_csv(tempDir+'/'+file, sep='\t')
    df = df.drop(['Unnamed: 0', 'Business Days'], axis=1)
    graphTitle = file.replace('_',' ').replace('.csv','').replace('ValuesComposit', 'Values Composite')
    fig = df.plot(kind='line', figsize=(8, 6), fontsize=12, title=graphTitle, legend=True).get_figure()
    fig.savefig(tempDir+'/'+file.replace(".csv", ".pdf"))
    plt.close('all')
    # clear_output(wait=False)
    
```

```python
# ==============================================================================
# Plotting Models Composite DataFrames
# ==============================================================================

count = 0
for file in ModelsComposit_fileList:
    count += 1
    print(count, ': '+tempDir+'/'+file)
    df = pd.read_csv(tempDir+'/'+file, sep='\t')
    df = df.drop(['Unnamed: 0', 'Business Days'], axis=1)
    graphTitle = file.replace('_',' ').replace('.csv','').replace('ModelsComposit', 'Models Composite')
    fig = df.plot(kind='line', figsize=(8, 6), fontsize=12, title=graphTitle, legend=True).get_figure()
    fig.savefig(tempDir+'/'+file.replace(".csv", ".pdf"))
    plt.close('all')
    # clear_output(wait=False)
```
